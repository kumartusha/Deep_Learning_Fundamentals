# Single Layer Perceptron

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv("placement.csv")
dataset.head()
```

```python
plt.figure(figsize=(7,4))
sns.scatterplot(x = "cgpa", y = "iq", data = dataset, hue = "placement")
plt.show()
```

```python
x = dataset.drop("placement", axis = 1)
y = dataset["placement"]
```

```python
#  Train test split.
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=20)
```

```python
from sklearn.linear_model import Perceptron
```

```python
#  Not work well in case of the perceptron.  
pr = Perceptron(alpha=1)    # Give only the single line for the single perceptron.
```

```python
pr.fit(x_train, y_train)
```

```python
#  Check the accuracy of the model.

training_data_pred = pr.predict(x_train)
training_data_accu = accuracy_score(training_data_pred , y_train)
print("The accuracy for the training data are:- ", training_data_accu * 100)

#  Check the accuracy of the model.

testing_data_pred = pr.predict(x_test)
testing_data_accu = accuracy_score(testing_data_pred , y_test)
print("The accuracy for the testing data are:- ", testing_data_accu * 100)

#  Other method for check the accuracy for the model.
print(pr.score(x_train, y_train))
print(pr.score(x_test, y_test))
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf = pr)
plt.show()
```

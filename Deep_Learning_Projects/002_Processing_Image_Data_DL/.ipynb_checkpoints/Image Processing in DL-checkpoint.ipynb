## How to Resize the image using the pillow library.

```python
from PIL import Image
```

```python
img = Image.open('dog.webp')
img_resized = img.resize((1000,1000))
img_resized
```

```python
print(img_resized)
```

```python
img_resized.save('dog__resized.jpg')
```

```python

img_resi = mpimg.imread('dog__resized.jpg')
img_res_plot = plt.imshow(img_resi)
plt.show()
```

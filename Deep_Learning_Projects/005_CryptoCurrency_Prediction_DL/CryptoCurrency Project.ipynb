## Using the LSTM ->> Long Short Term Memory.

In the time series we need to store the previous information and for doing this we have the RNN but RNN
is only store the limited data so we need the LSTM for supporting the RNN functionality.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
scaled_data = scaler.fit_transform(closing_price[["Close"]].dropna())
```

```python
scaled_data.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, LSTM
```

```python
scaled_data.shape
```

```python
# Prepared the data for the LSTM.

x_data = []
y_data = []

base_days = 100                                                                                             
for i in range(base_days, len(scaled_data)):
    x_data.append(scaled_data[i - base_days: i])
    y_data.append(scaled_data[i])

x_data, y_data = np.array(x_data), np.array(y_data)
```

```python
x_data.size
```

```python
x_data.shape
```

```python
# Split into the train and test sets.
train_size = int(len(x_data) * 0.9)
x_train , y_train = x_data[:train_size], y_data[: train_size]
x_test, y_test = x_data[train_size: ], y_data[train_size:]

print(train_size)
# print(x_train)
print(y_train)
```

## Building the LSTM Model

```python
model = Sequential([
    LSTM(128, return_sequences=True, input_shape = (x_train.shape[1],1)),
    LSTM(64, return_sequences=False),
    Dense(25),
    Dense(1)
])
```

```python
#  Compile the Mode.
model.compile(optimizer="adam", loss = "mean_squared_error")
```

```python
model.summary()
```

```python
model.fit(x_train, y_train, batch_size=5, epochs = 10)
```

```python
predictions = model.predict(x_test)

inv_predictions = scaler.inverse_transform(predictions)
inv_y_test = scaler.inverse_transform(y_test)
```

```python
print(predictions)
print(inv_predictions)
```

```python
inv_y_test
```

```python
# Plotting the Graph for the actual and the predicted Value.
plotting_data = pd.DataFrame({
    'Original' : inv_y_test.flatten(),
    'Prediction': inv_predictions.flatten(),
}, index = closing_price.index[train_size + base_days:]
                            )

#  Plot the Graph for the Closing Price.
plt.figure(figsize=(10,6))
plt.plot(plotting_data.Original, color = "Blue", linewidth = 1, label = "Original")
plt.plot(plotting_data.Prediction, color = "green", linewidth = 1, label = "Prediction")

plt.title("Prediction Vs Actual Close Price")
plt.xlabel("Years", fontsize = 14)
plt.ylabel("Close Price", fontsize = 14)
plt.grid(alpha = 0.9)
plt.legend(fontsize = 12)
plt.show()
```

```python
#  Predict future days.
last_100 = scaled_data[-100:].reshape(1,-1,1)
future_predictions = []

for _ in range(10):
    next_days = model.predict(last_100)
    future_predictions.append(scaler.inverse_transform(next_days))
    last_100 = np.append(last_100[:, 1: , :], next_days.reshape(1,1,-1), axis = 1)
```
